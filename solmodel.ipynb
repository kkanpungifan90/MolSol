```python
plt.scatter(df['Molecular Weight'], df['ESOL predicted log solubility in mols per litre'])

plt.xlabel('Molecular Weight')  # Setting the label for the x-axis
plt.ylabel('ESOL predicted log solubility in mols per litre')  # Setting the label for the y-axis
plt.title('Molecular Weight vs. ESOL Predicted Log Solubility')  # Setting the title for the scatter plot

plt.show()
```

```python
# Assuming 'df' is your DataFrame
features = ['Molecular Weight', 'Number of H-Bond Donors', 'Number of Rotatable Bonds', 'Polar Surface Area']
X = df[features]  # Independent variables
Y = df['measured log solubility in mols per litre']  # Dependent variable
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE):', mean_squared_error(Y_test, Y_pred))
print('Coefficient of determination (R^2):', r2_score(Y_test, Y_pred))
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(x=Y_test, y=Y_pred, c="#0000FF", alpha=0.3)

# Add trendline
z = np.polyfit(Y_test, Y_pred, 1)
p = np.poly1d(z)
plt.plot(Y_test, p(Y_test), "#F8766D")

plt.xlabel('Experimental LogS')
plt.ylabel('Predicted LogS')
plt.title('Experimental vs. Predicted LogS')

plt.show()
```
